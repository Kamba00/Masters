```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
r=1
Cd=0
x = []; y = [];
xc = []; yc = [];

for i in range(N):
    xi = np.random.uniform(-1,1)
    yi = np.random.uniform(-1,1)
    x.append(xi)
    y.append(yi)
    if ((xi**2 + yi**2) <= r ):
        xc.append(xi)
        yc.append(yi)
        Cd+=1
        
print(len(y))
plt.scatter(x,y)
plt.scatter(xc,yc)
plt.gca().set_aspect('equal')
```

```python
N = 1000
m=-1
c=1
x = [];y = [];
xs = [];ys = [];

for i in range(N):
    xi = np.random.uniform(0,1)
    yi = np.random.uniform(0,1)
    x.append(xi)
    y.append(yi)
    if (m*xi + c) >= yi:  
        xs.append(xi)
        ys.append(yi)
plt.scatter(x,y)
plt.scatter(xs,ys)
```

```python
def fx(x):
    n = np.exp(-x**2)
    return n
```

```python
p=[]
def monte(a=-3, b=3, N=1000000):   
    integral = 0
    for i in range(N):
        p.append(np.random.uniform(a,b))
    
    for x in p:
        integral+=fx(x)

    return ((b-a)/N)*integral

monte()

```
