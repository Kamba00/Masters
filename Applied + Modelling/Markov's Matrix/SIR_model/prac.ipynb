```python
def SIR(y,t,a,b):
    S ,I ,R = y
    dSdt = -a*S*I
    dIdt = a*S*I - b*I
    dRdt = b*I
    return ([dSdt , dIdt , dRdt])

t = np.linspace(0,100,100)
S0 = 0.9; I0 =.1; R0 = 0.0;
a = 0.35 # infection rate
b = 0.1 # recovery

ans = odeint(SIR, [S0, I0, R0], t, args = (a,b,))
ans= np.array(ans)
```

```python
plt.plot(t, ans[:,0],label = "S")#Susceptible
plt.plot(t, ans[:,1],label = "I")#Infected
plt.plot(t, ans[:,2],label = "R")#removed
plt.legend()
```
