```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as ran
```

```python
nodes = ["Win","Lose"]
trans_matrix = np.array ([[0.5,0.5],[0.5,0.5]])
```

```python

def gamble():
    start_money = 500
    bet_size = 100
    max_win = 1000
    step = 100
    simul = []
    bet_range = np.arange(bet_size,max_win,step)
    while start_money in bet_range:
        bet = 100 * ran.randrange(-1,3,step=2)
        start_money += bet
        simul.append(start_money)
    return simul
```

```python
plt.plot(gamble(),marker="o")
```
